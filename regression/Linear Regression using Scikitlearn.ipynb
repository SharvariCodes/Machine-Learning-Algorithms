# <font color='light blue'> **Linear Regression using Scikitlearn**

# <font color='gree'> **Importing the Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax
```

# <font color='gree'> **Loading the Data**

```python
data = pd.read_csv('data_for_lr.csv')
```

```python
data.isnull()
```

# <font color='gree'> **Data Preprocessing**



```python
data.isnull().sum()
```

```python
data= data.dropna()
print("Shape of the data: ", data.shape)
```

```python
data.info()
```

# <font color='gree'> **Splitting the Data**

```python
train_input = np.array(data.x[0:500]).reshape(500, 1)
train_output = np.array(data.y[0:500]).reshape(500, 1)


test_input = np.array(data.x[500:700]).reshape(199, 1)
test_output = np.array(data.y[500:700]).reshape(199, 1)
```

```python
print(f"Training data shape: {train_input.shape}, {train_output.shape}")
print(f"Testing data shape: {test_input.shape}, {test_output.shape}")
```

# <font color='gree'> **Linear Regression**

### **Training**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(train_input, train_output)
```

```python
lr_model.coef_
```

```python
lr_model.intercept_
```

### **Testing**

```python
test_predictions = lr_model.predict(test_input)
```

```python
plt.plot(test_input, test_predictions, "+", color = 'green')
plt.plot(test_input, test_output, ".", color = 'red')
plt.title('Performance Testing')
plt.xlabel('Input')
plt.ylabel('Output/Predictions')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
cost = mean_squared_error(test_output, test_predictions)
cost
```
