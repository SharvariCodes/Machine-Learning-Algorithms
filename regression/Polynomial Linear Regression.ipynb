# <font color='gree'> **Polynomial Linear Regression**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Position_Salaries.csv')
```

```python
x = data.iloc[:, 1:-1].values
y = data.iloc[:, -1].values
```

```python
x
```

# <font color='gree'> **Linear Regression**


```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x, y)
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_reg.predict(x), color = 'blue')
plt.title('Linear Regression')
plt.xlabel('Input')
plt.ylabel('Predicitions')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(x)
```

```python
x_poly
```

```python
poly_linear_regression = LinearRegression()
poly_linear_regression.fit(x_poly, y)
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, poly_linear_regression.predict(x_poly), color = 'blue')
plt.title('Linear Regression')
plt.xlabel('Input')
plt.ylabel('Predicitions')
plt.show()
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, poly_linear_regression.predict(x_poly), color = 'blue')
plt.title('Linear Regression')
plt.xlabel('Input')
plt.ylabel('Predicitions')
plt.show()
```
